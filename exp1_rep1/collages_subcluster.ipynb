```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_pickle('df_cluster_img.pkl')
```

```python
# Define a function to plot a collage of random images
def plot_random_images(images, title, cmap, channel):
    num_images = len(images)
    fig, axes = plt.subplots(4, 4, figsize=(12, 12))
    fig.suptitle(title)
    for i in range(4):
        for j in range(4):
            if i * 4 + j < num_images:
                axes[i, j].imshow(images[i * 4 + j][:,:,channel], cmap=cmap)
                axes[i, j].axis('off')
            else:
                axes[i, j].axis('off')
    plt.savefig('cluster_imgs.png')
    plt.show()

# Iterate over each group
# Select the middle z-stack of each image
images = [image for image in df['intensity_image'].apply(lambda x: x[x.shape[0] // 2, :, :, :])]

# Select 16 random images if available, else select all
i_selected_images = np.random.choice(np.arange(0, len(images)), size=16, replace=False)
selected_images_2 = [images[i] for i in i_selected_images] 

# Plot the collage
plot_random_images(selected_images_2, f'Random images of cluster ', 'Blues', 0)
#plot_random_images(selected_images_2, f'Random Images ', 'Greens', 1)

```

```python
# Define a function to plot a collage of random images
def plot_random_images(images, title, cmap, channel):
    cmaps = ['Blues', 'Greens', 'Reds']
    num_images = len(images)
    fig, axes = plt.subplots(4, 3, figsize=(12, 12))
    fig.suptitle(title)
    for i in range(4):
        for j in range(3):
            if i * 4 + j < num_images:
                axes[i, j].imshow(images[i * 4][:,:,j], cmap=cmaps[j])
                axes[i, j].axis('off')
            else:
                axes[i, j].axis('off')
    plt.savefig('cluster_imgs_color.png')
    plt.show()

# Select the middle z-stack of each image
images = [image for image in df['intensity_image'].apply(lambda x: x[x.shape[0] // 2, :, :, :])]

# Select 16 random images if available, else select all
i_selected_images = np.random.choice(np.arange(0, len(images)), size=16, replace=False)
selected_images_2 = [images[i] for i in i_selected_images] 

# Plot the collage
plot_random_images(selected_images_2, f'Cluster Images', 'Blues', 0)

```
